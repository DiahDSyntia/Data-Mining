```python
import pickle
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold, cross_validate
```

# baca dataset

```python
data = pd.read_csv('https://raw.githubusercontent.com/DiahDSyntia/datamining-tugasakhir/main/indian_liver_patient.csv', sep=',')
data.head()
```

# Pemisahan Data

```python
#ambil label setiap baris
label = data.iloc[:,-1:].values.ravel()
label
```

```python
#ambil feature data
feature = data.iloc[:,:-1]
feature.values
```

```python
#pembagian data latih dan uji sebanyak 80% dan latih 20% untuk data set
X_train, X_test, y_train, y_test = train_test_split(feature, label, test_size = 0.3)
```

```python
print("Jumlah Data Latih:", len(X_train))
print("Jumlah Data Uji:", len(X_test))
```

# Pemodelan Dengan KNN

```python
# Create KNN classifier
clf = KNeighborsClassifier(n_neighbors = 3)
# Fit the classifier to the data
clf.fit(X_train, y_train)
```

# Prediksi Hasil

```python
pred = clf.predict(X_test)
pred
```

```python
cols = {
    'Y True' : y_test,
    'Predict' : pred,
}

print("Prediksi yang dihasilkan")
pd.DataFrame(cols)
```

# Pengukuran Permormance

```python
#fold = 10
kfold = KFold(n_splits=10)
scores = ['accuracy', 'precision', 'recall']
results = cross_validate(clf, X_test, y_test, cv=kfold, scoring=scores, return_train_score=True)
result_data = pd.DataFrame(results)
result_data
```

```python
print("PERFORMANCE MODEL KNN")

accuracy = result_data.test_accuracy.mean() * 100
precision = result_data.test_precision.mean() * 100
recall = result_data.test_recall.mean() * 100

print("Accuracy: %0.2f" % accuracy, "%")
print("Precision: %0.2f" % precision, "%")
print("Recall: %0.2f" % recall, "%")
```
